```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data.csv')
```

```python
df.head(5)
```

```python
df.info()
```

## Data Cleaning

```python
# fillna
# dropna
df.dropna(how='any', inplace=True)
```

```python
len(df)
```

```python
df.diagnosis.value_counts()
```

```python
df.diagnosis.value_counts()/len(df)*100
```

## Splitting the data into X & Y

```python
X = df.drop(['id', 'diagnosis'], axis=1)
```

```python
Y = df.diagnosis.values
```

## **Train Test Split**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test= train_test_split(X, Y, test_size=0.25)
```

```python
len(X_train)
```

```python
len(X_test)
```

## Feature Scaling 

```python
from sklearn.preprocessing import StandardScaler

sc= StandardScaler()
X_train_sc = sc.fit_transform(X_train)
X_test_sc = sc.transform(X_test)
```

```python
X_train_sc
```

```python
X_test_sc
```

## Model Building 

```python
from sklearn.neighbors import KNeighborsClassifier

# Intitiate the Classifier
model_knn = KNeighborsClassifier()

# Passing the data to classifier
model_knn.fit(X_train_sc, Y_train)
```

**In a Jupyter environment, please rerun this cell to show the HTML representation or trust the notebook.
On GitHub, the HTML representation is unable to render, please try loading this page with nbviewer.org**.

```python
Y_pred_knn = model_knn.predict(X_test_sc)
```

```python
Y_pred_knn
```

```python
Y_test
```

```python
# Classification metrics to check how the model behaves

from sklearn.metrics import accuracy_score

print(round(accuracy_score(Y_test, Y_pred_knn)*100,2))
```

## Model Building - DT

```python

from sklearn.tree import DecisionTreeClassifier

# Initiate the classifier
model_dt = DecisionTreeClassifier()

# Passing the data to classifier
model_dt.fit(X_train_sc, Y_train)
```
